# Somme de variables aléatoires indépendantes et loi des grands nombres

On va illuster la loi des grands nombres, qui dit que si $(X_i)_{i\geq 0}$ est une suite de variables aléatoires
i.i.d. qui admettent une espérance finie, alors 
$$
\frac{1}{n} \sum_{i=1}^n X_i  \text{  E(X_1)  n\to+\infty\, ,
$$
en un sens précis (donné dans l'énoncé du théorème).

On définit une fonction générique, qui prend en argument une fonction $X()$ et un entier $n$, et qui dessine le graphe de $( \frac{1}{m} \sum_{i=1}^m X_i )_{1\leq m\leq n}$

```python
def LGN(n,X):
    S=[X()]
    for m in range(2,n):
        S.append(S[m-2]*(m-1)/m+X()/m)
    plt.plot(S)
```

Quelques exemples de loi avec lesquelles on va tester la fonction LGN.

```python
p=0.5
def Ber():
    if rd.random()<p:
        return 1
    else:
        return 0
    
for i in range(3):
    LGN(1000,Ber)
plt.plot([0,1000],[p,p],'--',color='black')
plt.show()
```

```python
p=0.3
for i in range(3):
    LGN(5000,Ber)
plt.plot([0,5000],[p,p],'--',color='black')
plt.show()
```

```python
t=0.5
def Geom():
    r=1
    while rd.random()>t:
        r+=1
    return r

t=0.5
for i in range(3):
    LGN(5000,Geom)
plt.plot([0,5000],[1/t,1/t],'--',color='black')
plt.show()
```

```python
l=0.5
def Expo():
    return - log(rd.random())/l

for i in range(3):
    LGN(5000,Expo)
plt.plot([0,5000],[1/l,1/l],'--',color='black')
plt.show()
```

```python
for i in range(3):
    LGN(1000,rd.random)
plt.plot([0,1000],[0.5,0.5],'--',color='black')
plt.show()
```

## Une loi qui n'admet pas d'espérance: la loi de Cauchy

```python
def Cauchy():
    return tan(pi*(rd.random()-0.5))
```

```python
for i in range(3):
    LGN(1000,Cauchy)
plt.show()
```

```python
def AbsCauchy():
    return abs(Cauchy())
```

```python
for i in range(3):
    LGN(10000,AbsCauchy)
plt.plot([t for t in range(10000)],[log(1+t) for t in range(10000)],'--',color='black')
plt.show()
for i in range(3):
    LGN(1000000,AbsCauchy)
plt.plot([t for t in range(1000000)],[log(1+t) for t in range(1000000)],'--',color='black')
plt.show()
```
